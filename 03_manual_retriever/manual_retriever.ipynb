# Manual Retrieval in RAG

In Retrieval-Augmented Generation (RAG), retrieval is the process of finding relevant documents before sending them to an LLM.

In this notebook, we will:

1. Create a small document collection.
2. Convert documents into embeddings.
3. Convert a query into an embedding.
4. Calculate cosine similarity manually.
5. Retrieve the most relevant document.

This demonstrates the core retrieval mechanism used in RAG systems.

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer(
    'sentence-transformers/all-MiniLM-L6-v2'
)
```

```python
documents = [
    "Python is a popular programming language.",
    "Machine learning enables computers to learn from data.",
    "RAG combines retrieval and generation.",
    "Deep learning uses neural networks.",
    "OpenCV is used for computer vision tasks."
]
```

```python
document_embeddings = model.encode(documents)

print(document_embeddings.shape)
```

```python
query = "What is retrieval augmented generation?"
```

```python
query_embedding = model.encode([query])

print(query_embedding.shape)
```

```python
scores = cosine_similarity(
    query_embedding,
    document_embeddings
)

print(scores)
```

```python
best_index = np.argmax(scores)

print("Best Index:", best_index)
```

```python
print("Query:")
print(query)

print("\nRetrieved Document:")
print(documents[best_index])
```

```python
k = 3

top_k_indices = np.argsort(scores[0])[::-1][:k]

top_k_indices
```

```python
for rank, idx in enumerate(top_k_indices, start=1):
    print(f"Rank {rank}")
    print(f"Score: {scores[0][idx]:.4f}")
    print(f"Document: {documents[idx]}")
    print("-" * 50)
```

```python
new_query = "Tell me about neural networks"

query_embedding = model.encode([new_query])

scores = cosine_similarity(
    query_embedding,
    document_embeddings
)

best_index = np.argmax(scores)

print(documents[best_index])
```
