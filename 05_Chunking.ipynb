# Chunking in RAG

Chunking is the process of breaking large documents into smaller pieces called chunks.

Why do we need chunking?

- Better retrieval accuracy
- Smaller and more meaningful embeddings
- More relevant context sent to the LLM
- Reduced token usage

Without chunking, an entire document is converted into a single embedding.

With chunking, multiple embeddings are created, allowing the retrieval system to find the most relevant section of the document.

```python
document = """
Python is a popular programming language.

Machine learning is a branch of artificial intelligence.

Deep learning uses neural networks with many layers.

RAG combines retrieval and generation.

OpenCV is used for computer vision applications.
"""
```

```python
def charecter_chunk(text,chunk_size=50):
    chunks = []
    for i in range(0,len(text),chunk_size):
        chunks.append(text[i:i+chunk_size])
    return chunks
```

```python
charecter_chunks = charecter_chunk(
    document,
    chunk_size=50
)

charecter_chunks
```

```python
for i, chunk in enumerate(charecter_chunks, start=1):

    print(f"\nChunk {i}")
    print(chunk)
    print("-"*50)
```
