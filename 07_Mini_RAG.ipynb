# Mini RAG System (From Scratch)

This notebook builds a complete Retrieval-Augmented Generation (RAG) system using:

- Sentence Transformers (embeddings)
- FAISS (vector database)
- Manual prompt construction
- Simple LLM simulation (later real LLM)

This is the foundation of all modern RAG systems like ChatGPT plugins, LangChain apps, and vector search assistants.

```python
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer(
    "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
documents = [
    "Python is a high-level programming language used for AI and web development.",
    "Machine learning allows computers to learn patterns from data.",
    "Deep learning uses neural networks with multiple layers.",
    "RAG combines retrieval systems with large language models.",
    "FAISS is a library for fast similarity search over vectors.",
    "OpenCV is used for image processing and computer vision tasks."
]
```

```python
chunks = documents  # already small, so we treat each sentence as chunk
```

```python
chunk_embeddings = model.encode(chunks)

chunk_embeddings = np.array(chunk_embeddings).astype('float32')

print(chunk_embeddings.shape)
```

```python
dimension = chunk_embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(chunk_embeddings)

print('Vectors in DB:', index.ntotal)
```

```python
def retrive(query,k=2):

    query_embedding = model.encode([query])
    query_embedding = np.array(query_embedding).astype('float32')

    distance,indices = index.search(query_embedding,k)

    results = [chunks[i] for i in indices[0]]

    return results
```

```python
query = 'What is FAISS used for?'

results = retrive(query)

results
```
