```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents = [
    "Python is a programming language",
    "TensorFlow is a deep learning framework",
    "CNN is used for image processing"
]

question = 'What is TensorFlow?'
```

```python
doc_embeddings = model.encode(documents)

question_embedding = model.encode(question)
```

```python
print("Document Embedding Shape:")
print(doc_embeddings.shape)

print("\nQuestion Embedding Shape:")
print(question_embedding.shape)
```

## Calculate Similarity

Cosine Similarity measures how close two vectors are.

Range:
- 1.0 → Exactly similar
- 0.0 → No similarity
- -1.0 → Opposite direction

Higher score = More relevant document

```python
similarities = cosine_similarity(
    [question_embedding],
    doc_embeddings
)

print(similarities)
```

```python
for i, score in enumerate(similarities[0]):
    print(f"Document {i+1}: {score:.4f}")
```

```python
best_match = similarities.argmax()

print("Best Document Index:", best_match)
print("\nRetrieved Document:")
print(documents[best_match])
```

```python
top_score = similarities[0][best_match]

print(f"Similarity Score: {top_score:.4f}")
```
